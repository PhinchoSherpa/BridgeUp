```python
import matplotlib.pyplot as plt 
import numpy as np 
import csv
import pandas as pd
```

```python
csvFile = open("sea_temp_anom_Dec_2002.csv", 'r') 
csvReader = csv.reader(csvFile, delimiter=',')
data_lists = list(csvReader)
```

```python
width=len(data_lists[0])
height=len(data_lists)
```

```python
grid = np.empty([height, width, 3], dtype=np.uint8)
```

```python
all_values = [] # We will fill in this empty list with all the data values as we go

for r, row in enumerate(data_lists):
    for c, value in enumerate(row):
        if float(value) < 99998.0:
            all_values.append(float(value)) # This is where we fill in the list with all data values
        if float(value) <= -5:
            grid[r, c] = [46, 0, 255]
        elif float(value) <= -4:
            grid[r, c] = [0, 135, 255]
        elif float(value) <= -3:
            grid[r, c] = [0, 221, 255]
        elif float(value) <= -2:
            grid[r, c] = [0, 255, 229]
        elif float(value) <= -1:
            grid[r, c] = [0, 255, 144]
        elif float(value) <= 0:
            grid[r, c] = [25, 255, 0]
        elif float(value) <= 1:
            grid[r, c] = [182, 255, 0]
        elif float(value) <= 2:
            grid[r, c] = [242, 255, 0]
        elif float(value) <= 3:
            grid[r, c] = [255, 174, 0]
        elif float(value) <= 4:
            grid[r, c] = [255, 0, 0]
        elif float(value) <= 5:
            grid[r, c] = [168, 1, 4]
        else:
            grid[r, c] = [0, 0, 0]
```

```python
print(min(all_values))
print(max(all_values))
```

```python
plt.title(" sea temperature anomaly")
plt.imshow(grid)
plt.axis('off')
plt.show()
```
