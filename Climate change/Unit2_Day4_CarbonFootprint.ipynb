# Calculating Carbon Footprint

In this notebook you'll learn about the input() function and find out how much CO2 each family generates.

**Reminder: To run the code in a Jupyter Notebook cell, click inside the cell and press "shift" and "return" on your keyboard at the same time.**

```python
num = input("What is your favorite number?")
print(num)
```

```python
# Import matplotlib
import matplotlib.pyplot as plt
```

```python
# make a blank list that will hold each value of CO2 used for driving, air, etc.
pounds_of_co2 = []
```

```python
# Calculate the amount of CO2 used per family by looking at the miles per year driven
# and the mileage per gallon of the vehicle.
amount_per_year = float(input("How many miles does your family drive per year? "))
miles_per_gallon = float(input("How many miles per gallon does your vehicle get? "))
total_co2_per_year = int(amount_per_year / miles_per_gallon) * 22
print("Your car uses " + str(total_co2_per_year) + " pounds of CO2 per year.\n")

#put the amount of CO2 used by cars into the list
pounds_of_co2.append(total_co2_per_year)
```

```python
# Calculate the amount of CO2 put into the air by flying.
amount_per_year = float(input("How many total miles does your family fly each year?"))
total_co2_per_year = amount_per_year * 0.9
print("Your family uses " + str(total_co2_per_year) + " pounds of CO2 per year due to air travel.\n")

#Add the total amount of CO2 from flying to the list.
pounds_of_co2.append(total_co2_per_year)
```

```python
# Calculate the amount of CO2 generated by the mass transit the family uses
amount_per_year = float(input("How many miles per year does your family travel on mass transit?"))
total_co2_per_year = amount_per_year * 0.5
print("Your family uses " + str(total_co2_per_year) + " pounds of CO2 per year due to mass transit.\n")

#Add the total amount of CO2 from mass transit to the list.
pounds_of_co2.append(total_co2_per_year)
```

```python
# Calculate the amount of CO2 generated by the taxis the family takes
amount_per_year = float(input("How many miles per year does your family travel in taxis?"))
total_co2_per_year = amount_per_year * 1.5
print("Your family uses " + str(total_co2_per_year) + " pounds of CO2 per year due to taxi travel.\n")

#Add the total amount of CO2 from taxis to the list.
pounds_of_co2.append(total_co2_per_year)
```

```python
# Calculate the amount of CO2 generated by electricity the family uses at home
amount_per_year = float(input("How many kilowatt hours of electricity does your household use?"))
total_co2_per_year = amount_per_year * 1.5
print("Your family uses " + str(total_co2_per_year) + " pounds of CO2 per year due to electricity.\n")

#Add the total amount of CO2 from electricity use to the list.
pounds_of_co2.append(total_co2_per_year)
```

```python
# Calculate the amount of CO2 generated by heating oil the family uses
amount_per_year = float(input("How many gallons of heating oil does your family use per year?"))
total_co2_per_year = amount_per_year * 22
print("Your family uses " + str(total_co2_per_year) + " pounds of CO2 per year due to heating oil.\n")

#Add the total amount of CO2 from heating oil use to the list.
pounds_of_co2.append(total_co2_per_year)
```

```python
# Calculate the amount of CO2 generated by natural gas the family uses
amount_per_year = float(input("How many therms per year does your family use for natural gas?"))
total_co2_per_year = amount_per_year * 11
print("Your family uses " + str(total_co2_per_year) + " pounds of CO2 per year due to natural gas.\n")

#Add the total amount of CO2 from natural gas use to the list.
pounds_of_co2.append(total_co2_per_year)
```

```python
# Calculate the amount of CO2 generated by propane the family uses
amount_per_year = float(input("How many gallons of propane does your family use per year?"))
total_co2_per_year = amount_per_year * 13
print("Your family uses " + str(total_co2_per_year) + " pounds of CO2 per year due to propane useage.\n")

#Add the total amount of CO2 from propane gas use to the list.
pounds_of_co2.append(total_co2_per_year)
```

```python
# Calculate and print the total amount of CO2 generated by the family
grand_total = 0
for amount in pounds_of_co2:
  grand_total = grand_total + amount
  
print("The grand total amount of CO2 used by this family is " + str(grand_total) + " pounds.")
```

```python
# Plot the family's CO2 use in each category
plt.clf()
plt.bar([1,2,3,4,5,6,7,8],pounds_of_co2, width = 0.5)
plt.xticks([1,2,3,4,5,6,7,8],("Car", "Plane", "Mass Transit", "Taxi", "Electricity", "Heating Oil", "Natural Gas", "Propane"), rotation=50)
plt.title("CO2 use by family")
plt.ylabel("Pounds of CO2")
plt.savefig("carbonfootprint.png")
plt.show()
```
