```python
#Importing libraries
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Opening each csv file and assigning it a variable
hawaii=pd.read_csv("Hawaii_Bleaching.csv")
northern=pd.read_csv("Northern_Great_Barrier_Reef_Bleaching.csv")
fiji=pd.read_csv("Fiji_Bleaching.csv")
florida=pd.read_csv("Florida_Keys_Bleaching.csv")
galapagos=pd.read_csv("Galapagos_Islands_Bleaching.csv")
palau=pd.read_csv("Palau_Bleaching.csv")
```

```python
#Assigning each lat and long a variable.
latH=hawaii["Latitude"]
longH=hawaii["Longitude"]
latN=northern["Latitude"]
longN=northern["Longitude"]
latF=fiji["Latitude"]
longF=fiji["Longitude"]
latFl=florida["Latitude"]
longFl=florida["Longitude"]
latG=galapagos["Latitude"]
longG=galapagos["Longitude"]
latP=palau["Latitude"]
longP=palau["Longitude"]
```

```python
image = plt.imread("earth.jpg")#background image
fig, ax = plt.subplots() #for plotting
ax.imshow(image, extent = [-180,185,-90,90])
plt.xlim(-180,185)
#Assigning the axis and color
plt.scatter(longH,latH,color="orange")
plt.scatter(longN,latN,color="cyan")
plt.scatter(longF,latF,color="gray")
plt.scatter(longFl,latFl,color="blue")
plt.scatter(longG,latG,color="lime")
plt.scatter(longP,latP,color="purple")
plt.legend(labels=["Hawaii","Northern Great Barrier Reef","Fiji","Florida Keys","Galapagos Islands","Palau"],bbox_to_anchor=(1,1))#labels outside the map
plt.title("World Map")
plt.savefig("SST Per Year.png",bbox_inches='tight')
plt.show()
```
