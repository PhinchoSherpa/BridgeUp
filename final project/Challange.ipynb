```python
groupby_year = northern["Sea_Surface_Temperature"].groupby(northern["Year"]).mean()
groupby_year.to_csv("Sea_Temperature_mean.csv")
```

```python
df2=pd.read_csv("Sea_Temperature_mean.csv")
```

```python
plt.plot(df2["Year"],df2["Sea Temperature Mean"])
plt.xlabel("Year")
plt.ylabel("Temperature")
plt.title("Northern Barrier Sea Temperature")
plt.savefig("Northern Barrier Temperature.png",bbox_inches='tight')
plt.show()
```
