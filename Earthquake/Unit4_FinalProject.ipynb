### Let's add some color (and functions)
Next, you'll need to color code your points based on one of the characteristics in the data frame. 

First, create a function that takes the name of a column and a number of colors and returns list of intermediate values based on the number of colors you specify (these values will be the boundaries for conditionals later on).

```python
colorList = []#add in this list later on
for value in df["EQ_PRIMARY"]: #magnitude
    if value <=4.4:
        colorList.append([126/255,55/255,232/255])#purple
    elif value<=5.0:
        colorList.append([22/255,117/255,8/255])#green
    elif value<=5.5:
        colorList.append([18/255,55/255,150/255])#dark blue
    elif value<=6.0:
        colorList.append([216/255,199/255,8/255])#yellow
    elif value<=6.5:
        colorList.append([196/255,19/255,9/255])#red
    else:
        colorList.append([59/255,219/255,247/255])#blue
    
```

```python
def Pdata(df,column1,column2): #creating a new function
    lat=df[column1]
    long=df[column2]
    plt.scatter(long,lat,color=colorList) 
```

```python
image = plt.imread("earth.jpg")
fig, ax = plt.subplots()
ax.imshow(image, extent = [-180,180,-90,90])
Pdata(df,"LATITUDE","LONGITUDE")
plt.xlim(-100,-70)
plt.ylim(0,25)
plt.show()
```
